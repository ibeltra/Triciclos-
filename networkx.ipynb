```python
import networkx as nx
from networkx.drawing.nx_pylab import draw
```

```python
def read_graph(filename):
    graph = nx.Graph()
    with open(filename) as f:
        for line in f:
            graph.add_edge(*line.strip().split(','))
    return graph
        
```

```python
g0 = read_graph("g0.txt")
draw(g0, with_labels=True)
```

```python
g1 = read_graph("g1.txt")
draw(g1, with_labels=True)
```
